```python
import numpy as np
import matplotlib.pyplot as plt

ph=np.arange(6,10, step=0.01)

precip_limited=5*10**(11-2*ph)*10**(ph-9.5)
fe2_limited=1.0E-4*10**(ph-9.5)

feoh=np.minimum(precip_limited, fe2_limited)

plt.plot(ph, feoh, color='black')
plt.plot(ph, precip_limited, color='red', linestyle='--', label='Precipitation-Limited')
plt.plot(ph, fe2_limited, color='blue', linestyle='--',label='[Fe2+] - limited' )

plt.xlabel('pH')
plt.ylabel('[FeOH+] (M)')
plt.ylim(1.0E-9, 1.0E-4)
plt.yscale('log')
plt.legend()
###What depth is OD=1 (ignoring slant angle for now)?
print(np.log10(np.exp(1))/(3.55E2*2.0E-6))
```
